```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='age', y='salary', color='blue', edgecolor='black', alpha=0.5)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='healthy_eating', y='active_lifestyle', color='blue', edgecolor='black', alpha=0.5)

plt.title('Healthy Eating vs Active Lifestyle')
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='healthy_eating', y='salary', color='blue', edgecolor='black', alpha=0.5)

plt.title('Healthy Eating vs Salary')
plt.xlabel('Healthy Eating')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data = data, x='groups', edgecolor='linen', alpha=0.7)

plt.title('Count of Employees in Each Group')
plt.xlabel('Group')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data = data, x = 'salary', color='blue', edgecolor='black', alpha=0.7, bins=10)

plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```
